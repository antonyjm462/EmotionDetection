```python
datalist = []
for file in filelist:
  df = pd.read_csv(f'/content/drive/My Drive/Emotion-Detection/mod_dataset/{file}',lineterminator='\n')
  datalist.append(df)
```

```python
DF = pd.concat(datalist, ignore_index=True)
```

```python
DF.shape
```

```python
DF.Emotion.unique()
```

```python
texts = np.asarray(DF['Text'].values)
string = []
for text in texts:
    doc = str(text)
    doc = re.sub(pattern, ' ', doc)
    string.append(doc)
```

```python
Dataset = list(zip(string,DF['Emotion']))
DF = pd.DataFrame(data = Dataset, columns=['Text', 'Emotion'])
DF.to_csv("/content/drive/My Drive/Emotion-Detection/jira_dataset_final.csv", index=False, header=True)
```
